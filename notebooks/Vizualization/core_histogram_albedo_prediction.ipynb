## Core Workflow: Display Webmap from ArcGIS online 
Purpose: The roof albedo prediction for different years are stored as csv files. This notebook is used to create histogram for the albedo prediction from different years.
<br>

```python
import pandas as pd
import matplotlib.pyplot as plt
# import seaborn as sns
import os
import numpy as np

import warnings
warnings.filterwarnings('ignore')
```

### Histogram for BanWeiss prediction

```python
df = pd.read_csv('LosAngeles_final_albedo_dataset_key_vars_trimmed.csv')
df['ScoredLabels']=df['CALIB_SR']
x = df.ScoredLabels.dropna()
plt.figure(figsize=(12,12))
N, bins, patches = plt.hist(x,bins=[0, 0.05, 0.1, 0.15, 0.2, 0.25, 0.3, 0.35, 0.4, 0.45, 0.5, 0.55, 0.6, 0.65, 0.7, 0.75, 
                                    0.8, 0.85,0.9,0.95,1])

cmap = plt.get_cmap('jet')
low = cmap(0.5)
medium =cmap(0.2)
high = cmap(0.7)


for i in range(0,1):
    patches[i].set_facecolor('#d7191c')
for i in range(1,2):
    patches[i].set_facecolor('#ea633e')
for i in range(2,3):
    patches[i].set_facecolor('#fdae61')
for i in range(3,4):
    patches[i].set_facecolor('#fed690')
for i in range(4,5):
    patches[i].set_facecolor('#ffffbf')
for i in range(5,6):
    patches[i].set_facecolor('#d5ecd4')
for i in range(6,7):
    patches[i].set_facecolor('#c7dcef')
for i in range(7,8):
    patches[i].set_facecolor('#a2cbe2')
for i in range(8,9):
    patches[i].set_facecolor('#72b2d7')
for i in range(9,10):
    patches[i].set_facecolor('#4997c9')
for i in range(10,11):
    patches[i].set_facecolor('#2878b8')
for i in range(11,12):
    patches[i].set_facecolor('#0d57a1')
for i in range(12,18):
    patches[i].set_facecolor('#08306b')
plt.grid()    
plt.axis([0, 1, 0, 450000])
plt.xlabel("Predicted Albedo", fontsize=16)  
plt.ylabel("Count of Roofs", fontsize=16)
plt.axvline(x.mean(), color='#544f50', linestyle='dashed', linewidth=1)
plt.xticks(np.arange(0, 1.01, 0.05),fontsize=10)  
plt.yticks(np.arange(0, 500000, 50000),fontsize=14)
plt.text(0.15, 400000, "Mean: " + str(round(df.ScoredLabels.mean(),4)), style='italic',fontsize=14)
ax = plt.subplot(111)  
ax.spines["top"].set_visible(False)  
ax.spines["right"].set_visible(False)

plt.show()
```
